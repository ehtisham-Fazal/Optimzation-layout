```python
from gurobipy import *
```

```python
int_model= Model("Knapsack problem")
```

```python
w=[4,2,5,4,5,1,3,5]
v=[10,5,18,12,15,1,2,8]
c=15
N=len(w)
```

```python
x=int_model.addVars(N, vtype=GRB.BINARY, name="x")
```

```python
obj_fn=sum(v[i] * x[i] for i in range (N))
int_model.setObjective(obj_fn,GRB.MAXIMIZE)
```

```python
int_model.addConstr(sum(w[i] * x[i] for i in range(N)) <= c)
```

```python
int_model.setParam('OutputFlag',False)
```

```python
int_model.optimize()
```

```python
print(int_model.ObjVal)
```
